```python
import numpy as np
import itertools
import pandas as pd
import collections
```

```python
#Task 1

#Assinging values
M = 8
W = [2,1,3,5,6]
V = [5,2,4,13,14]

#Creating dictionary
dic = dict(zip(W,V))

combos = []
values = []
weights = []

#Find all possible combinations
for i in range(1,len(W)+1):
    perm = itertools.combinations(dic.keys(),i)
    combos += list(perm)

#Evaluate wights and values for all combos
for i in combos:
    counter = 0
    for l in i:
        counter += dic[l]
    values.append(counter)
    
for i in combos:
    weights.append(sum(i))

#Create DataFrame and find max value that saisfies condition
df = pd.DataFrame()
df['combos'] = combos
df['values'] = values
df['weights'] = weights

best_combo = df.iloc[df.loc[df['weights'] <= 8]['values'].idxmax()][0]
best_val = df.iloc[df.loc[df['weights'] <= 8]['values'].idxmax()][1]

print('The best weights to take are = ' + str(best_combo))
print('These give a value of = ' + str(best_val))
    
```

```python
#Task 2
#Assinging values
M = 8
W = [2,1,3,5,6]
V = [5,2,4,13,14]

#Finding value per wieght and sorting
val_per_weight = np.array(V)/np.array(W)
tup  = list(zip(val_per_weight, W))
tup_s = sorted(tup, key= lambda x: x[0], reverse=True)

#Adding in optimal order up to max weight
tot_w = 0
tot_v = []
weights = []
for i in tup_s:
    counter = 0
    while counter < i[1]:
        if tot_w == M:
            break
        else:
            tot_w += 1
            tot_v.append(i[0])
            weights.append(i[1])
        counter +=1

#Finding fractions of each weight added
fracts = [dict(collections.Counter(weights))[i]/i for i in dict(collections.Counter(weights)).keys()]

print('Optimal value = ' + str(sum(tot_v)))
print('Objects take are : ' + str(weights))
print('Fractions of each objects are: ' + str(fracts))

```
