```python
import scipy
from scipy.integrate import quad
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Task 1

def odes(x, t):
    #Constants
    b = 0.3
    y = 1/14
    N = 30000
    
    S = x[0]
    I = x[1]
    R = x[2]
    
    #Odes
    ds_dt = -(b*S*I)/N
    di_dt = ((b*S*I)/N)-y*I
    dr_dt = y*I
    
    return [ds_dt, di_dt, dr_dt]

#Initial conditions
x0 = [29999,1,0]

#Times
t = np.linspace(0,365,1000)

#Using odeint to solves odes
x = odeint(odes, x0, t)

S = x[:,0]
I = x[:,1]
R = x[:,2]

#Plotting results
plt.plot(t, S, label='S')
plt.plot(t, I, label='I')
plt.plot(t, R, label='R')
plt.legend()
plt.xlabel('Time (days)')
plt.ylabel('Number of people')
plt.title('SIR model')
```

```python
#Task 2

def odes(x, t):
    #Constants
    b = 0.3
    y = 1/14
    N = 30000
    m = 0.05
    
    S = x[0]
    I = x[1]
    R = x[2]
    D = x[3]
    
    #Modified Odes
    ds_dt = -(b*S*I)/N
    di_dt = ((b*S*I)/N)-y*I
    dr_dt = y*I - m*I
    dd_dt = m*I
    
    return [ds_dt, di_dt, dr_dt, dd_dt]


#Initial conditions
x0 = [29999,1,0,0]

#Times
t = np.linspace(0,365,1000)

#Using odeint to solves odes
x = odeint(odes, x0, t)

S = x[:,0]
I = x[:,1]
R = x[:,2]
D = x[:,3]

#Plotting results
plt.plot(t, S, label='S')
plt.plot(t, I, label='I')
plt.plot(t, R, label='R')
plt.plot(t, D, label='D')
plt.legend()
plt.xlabel('Time (days)')
plt.ylabel('Number of people')
plt.title('SIR model')
```

```python
#Task 3

def odes(x, t):
    #Constants
    b = 0.3
    y = 1/14
    N = 30000
    m = 0.05
    r = 0.15
    
    S = x[0]
    I = x[1]
    R = x[2]
    D = x[3]
    
    #Modified Odes
    ds_dt = -(b*S*I)/N + r*R
    di_dt = ((b*S*I)/N)-y*I
    dr_dt = y*I - m*I - r*R
    dd_dt = m*I
    
    return [ds_dt, di_dt, dr_dt, dd_dt]


#Initial conditions
x0 = [29999,1,0,0]

#Times
t = np.linspace(0,365,1000)

#Using odeint to solves odes
x = odeint(odes, x0, t)

S = x[:,0]
I = x[:,1]
R = x[:,2]
D = x[:,3]

#Plotting results
plt.plot(t, S, label='S')
plt.plot(t, I, label='I')
plt.plot(t, R, label='R')
plt.plot(t, D, label='D')
plt.legend()
plt.xlabel('Time (days)')
plt.ylabel('Number of people')
plt.title('SIR model')
```
